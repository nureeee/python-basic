```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import json
import pandas as pd
```

```python
url = 'http://www.bing.co.kr/brand/icecream'
```

```python
res = requests.get(url)
res
```

```python
data = res.content
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
ice_imgs = soup.select_one('#product-mobile-wrap')
ice_imgs
```

```python
ice_imgs_list = ice_imgs.select('img')
ice_imgs_list
```

```python
ice_imgs_list[0]['src']
```

```python
print('https:' + ice_imgs_list[0]['src'])
```

```python
if ice_imgs:
    for img in ice_imgs_list:
        img_url = img['src']
        img_url = "http://www.bing.co.kr/" + img_url 
        filename = img_url.split('/')[-1]
        print(img_url)
        urlretrieve(img_url, filename)
```
